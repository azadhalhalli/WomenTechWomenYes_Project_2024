```python
# Borough sütunundaki değerleri grupla ve toplam sayıyı hesapla
borough_counts = df['Borough'].value_counts()

# Çubuk grafiğini çiz
plt.bar(borough_counts.index, borough_counts.values, color='skyblue')
plt.xlabel('Borough')
plt.ylabel('Toplam Sayı')
plt.title('Borough Bazında Toplam Sayı')
plt.show()
```

```python
# Borough sütunundaki değerleri grupla ve yüzde oranını hesapla
borough_percentage = df['Borough'].value_counts(normalize=True) * 100

# Pasta grafiğini çiz
plt.pie(borough_percentage, labels=borough_percentage.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue', 'lightgreen', 'lightyellow', 'lightsalmon'])
plt.axis('equal')  # Daireyi mükemmel bir daire olarak ayarla
plt.title('Borough Dağılımı Yüzdeli Pasta Grafiği')
plt.show()
```
